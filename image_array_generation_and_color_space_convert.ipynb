```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
rgb = np.random.randint(0, 255, (256,256,3), np.uint8)
```

```python
plt.imshow(rgb)
```

```python
rgb = [[row, col,  int((row*col)/256)] for row in np.arange(255,-1,-1) for col in np.arange(255,-1,-1)]
```

```python
rgb = np.array(rgb, np.uint8).reshape(256,256,-1)
rgb.shape
```

```python
plt.imshow(rgb)
```

```python
value = 255
HSV = np.array([[h, s, value] \
    for s in np.arange(256) \
        for h in np.arange(180)], np.uint8).reshape(256, 180, 3)
```

```python
plt.imshow(HSV)
```

```python
RGB = cv2.cvtColor(HSV, cv2.COLOR_HSV2RGB)
plt.imshow(RGB)
```
