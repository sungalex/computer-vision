```python
# 1->2->1
%matplotlib inline

import keras
import numpy as np
import matplotlib.pyplot as plt

x_train = np.random.rand(100,1)
y_train = x_train * 2 + 1
```

```python
x = keras.layers.Input(shape=(1,))
h = keras.layers.Dense(2)(x)
y = keras.layers.Dense(1)(h)

model = keras.models.Model(x,y)
model.summary()
model.compile('sgd', 'mse')
hist = model.fit(x_train, y_train, batch_size=32, epochs=100, validation_split=0.2) # validation data 20%
```

```python
print(type(hist.history))
print(hist.history.keys())
print(len(hist.history['loss']))
```

```python
plt.plot(hist.history['loss'][10:], label="loss")
plt.plot(hist.history['val_loss'][10:], label="val_loss")
plt.xlabel("epoch")
plt.ylabel("loss")
plt.legend(loc="best")
plt.show()
```

```python
x_val = np.random.rand(100,1)
y_val = x_val * 2 + 1

model = keras.models.Model(x,y)
model.summary()
model.compile('sgd', 'mse')
hist = model.fit(x_train, y_train, batch_size=32, epochs=10, validation_data=(x_val, y_val))
```
